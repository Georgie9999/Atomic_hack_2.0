```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import shutil
```

```python
path_to_files = 'C:/Users/Пользователь/PycharmProjects/yolo_example'
i = 0
images = []
text = []
for f in os.listdir(os.path.join(path_to_files, 'dataset')):
    if i % 2 == 0: #image
        images.append(f)
    else:
        text.append(f)
    i += 1
```

```python
annotation = pd.DataFrame({'image_paths': images, 'text_paths': text})
```

```python
annotation
```

```python
train_part, val_part = train_test_split(annotation, test_size=0.25, random_state=42)
```

```python
train_part['part'] = 'train'
```

```python
train_part
```

```python
val_part['part'] = 'val'
```

```python
val_part
```

```python
annotation = pd.concat([train_part, val_part], axis=0)
```

```python
annotation
```

```python
annotation.to_csv('annotation.csv')
```

```python
annotation = pd.read_csv('annotation.csv', index_col=0)
annotation
```

```python
project_path = 'C:/Users/Пользователь/PycharmProjects/yolo_example'
```

```python
os.makedirs(os.path.join(project_path, 'data/images/train'), exist_ok=True)
os.makedirs(os.path.join(project_path, 'data/images/val'), exist_ok=True)
os.makedirs(os.path.join(project_path, 'data/labels/train'), exist_ok=True)
os.makedirs(os.path.join(project_path, 'data/labels/val'), exist_ok=True)
```

```python
train_images = list(annotation[annotation['part'] == 'train']['image_paths'])
val_images = list(annotation[annotation['part'] == 'val']['image_paths'])
```

```python
for img in train_images:
    shutil.copy(os.path.join(project_path, f'dataset/{img}'), os.path.join(project_path, 'data/images/train', img))
    an = img.replace('.jpg', '.txt')
    shutil.copy(os.path.join(project_path, f'dataset/{an}'), os.path.join(project_path, 'data/labels/train', an))

# Копирование валидационных изображений и аннотаций
for img in val_images:
    shutil.copy(os.path.join(project_path, f'dataset/{img}'), os.path.join(project_path, 'data/images/val', img))
    an = img.replace('.jpg', '.txt')
    shutil.copy(os.path.join(project_path, f'dataset/{an}'), os.path.join(project_path, 'data/labels/val', an))
```
